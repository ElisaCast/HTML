```python
# Descargar HTML
import requests
import time

url = "https://es.wikipedia.org/wiki/Leucocito"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Agrega un retardo de 10 segundos antes de la solicitud
time.sleep(10)
response = requests.get(url, headers=headers)

if response.status_code == 200:
    html_content = response.text
    print("HTML descargado exitosamente.")
else:
    print(f"No se pudo obtener el contenido HTML. Código de respuesta: {response.status_code}")
```

```python
# Transforma el HTML
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content,"html.parser")
print(soup)
```

```python
# BUSCAR TODAS LAS TABLAS
tables = soup.find_all("table")
tables
```

```python
import pandas as pd
# Encontrar la tabla que contiene información sobre los tipos de leucocitos
for index, table in enumerate(tables):
    if "Tipo" in str(table):
        table_index = index
        break

# Crear un DataFrame
df_tipos_leucocitos = pd.read_html(str(tables[table_index]))[0]

# Imprimir el DataFrame
print("Tabla de Tipos de Leucocitos:")
print(df_tipos_leucocitos)
```

```python
# Reemplazar caracteres no deseados y limpiar las filas
df_tipos_leucocitos = df_tipos_leucocitos.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df_tipos_leucocitos = df_tipos_leucocitos.applymap(lambda x: x.replace('$', '').replace(',', '') if isinstance(x, str) else x)

# Eliminar filas vacías o sin información
df_tipos_leucocitos = df_tipos_leucocitos.dropna(how='all')

# Imprimir el DataFrame después del procesamiento
print("DataFrame después del procesamiento:")
print(df_tipos_leucocitos)
```

```python
import sqlite3

# Nombre de la base de datos
nombre_db = "tipos_leucocitos.db"

# Conectar a la base de datos SQLite (si no existe, se creará)
conn = sqlite3.connect(nombre_db)
cursor = conn.cursor()

# Crear la tabla en la base de datos
cursor.execute('''
    CREATE TABLE IF NOT EXISTS tipos_leucocitos (
        Tipo TEXT,
        Apariencia_microscopica TEXT,
        Diagrama TEXT,
        Porcentaje_aproximado_adultos REAL,
        Diametro_um REAL,
        Principal_objetivo TEXT,
        Nucleo TEXT,
        Granulos TEXT,
        Vida_media TEXT
    )
''')

# Insertar los valores en la tabla
for _, row in df_tipos_leucocitos.iterrows():
    cursor.execute('''
        INSERT INTO tipos_leucocitos VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', tuple(row))

# Confirmar los cambios y cerrar la conexión
conn.commit()
conn.close()

print("Datos almacenados en la base de datos SQLite.")
```
